6. We can also compress images by modifying their pixels. Create a new image half the size of your original, using every other pixel. 

You can make it even more pixelated by choosing every third, fourth, fifth pixel, etc.  

```python
#every four pixles 
compressed_image1 = plt.imread("otter.jpg")
compressed_image1 = compressed_image1[::4, ::4]
imgplot = plt.imshow(compressed_image1)
```

```python
#every 8 pixles 
compressed_image2 = plt.imread("otter.jpg")
compressed_image2 = compressed_image2[::8, ::8]
imgplot = plt.imshow(compressed_image2)
```

7. We can also flip an image 90 degrees left or right by modifying the pixels of the image. Create a new image that is a 90 degree rotation counterclockwise of the original.

```python
#saves the array of image as "lst"
lst = image
#creates emtpy list 
lst2 = []
#from 0 to the width of the image lst[0]) append a new list inside the other list (2d array)
#width of "lst" becomes the length (#of lists) of "lst2"
for i in range(len(lst[0])):
    lst2.append([])
#from 0 to the length of the image ch = 0 and as long as ch is under 400 (the width 'lst') append a pixel ('o,o,o') to the row (the width)
for i in range(len(lst2)):
    ch = 0 
    while ch <= 399:
        lst2[i].append([0,0,0])
        ch += 1 
#for each unit of width cyle through each unit of length the width (i) of og image changes to the length (j) of 'lst2' and vice versa
for i in range(len(lst2[0])):
    for j in range(len(lst2)):
        lst2[j][i] = image[i][j]

#print values of height and width 
height = len(lst2)
width = len(lst2[0])

#convert the array of 'lst2' into pixes/image
imgplot = plt.imshow(lst2)

```

8. This is just for fun, but we can look at the red, green, and blue pixels of your image individually using array slicing. You may need to `pip install numpy`.

```python
import numpy as np 

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,8))
 
for c, ax in zip(range(3), axs):
    rgb_img = np.zeros(image.shape, dtype="uint8")
    rgb_img[:,:,c] = image[:,:,c]
    ax.imshow(rgb_img)
    ax.set_axis_off()
```
